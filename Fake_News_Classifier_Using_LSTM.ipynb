```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
### Vocabulary size
voc_size=5000
```

#Onehot Representation

```python
messages=X.copy()
```

```python
messages.reset_index(inplace=True)

# I used reset_index because I had dropped NaN values from the X.
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
### Dataset Preprocessing
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    print(i)
    review = re.sub('[^a-zA-Z]', ' ', messages['title'][i])
    review = review.lower()
    review = review.split()

    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
onehot_repr=[one_hot(words,voc_size)for words in corpus]
onehot_repr

# In one hot representation, we basically require the index of words based on the vocabulary size.
```

#Embedding Representation

```python
sent_length=20
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)

# If length of the sentences are different then we use "pad_sequences" to make all sentences of equal length. First we have fixed the length of the sentence. If we are using the padding parameter as "pre" then it will add 0 in the begining and if we use "post" then it wil add 0 in the end to make the sentences of equal length.
```

```python
#This is what we require as input to give our embedding layer.
```

```python
embedded_docs[0]
```

```python
## Creating model
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
# In embedding layer, it takes the input and converts it into specific number of features or vectors. Based on that number of features it give the output.
# Here I have taken it as 40, even we can take 100 features too.
# Then first I have created a sequencial layer.
# In the second layer, I am adding an embedding layer. Then I will pass the output of this embedding layer to LSTM layer. Here I have taken one LSTM layer with 100 neurons.
# Then I have added a dense layer, because it is a classification problem.
```

```python
len(embedded_docs),y.shape
```

```python
import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
X_final.shape,y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

#Model Training

```python
### Finally Training
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
from tensorflow.keras.layers import Dropout
## Creating model
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```
